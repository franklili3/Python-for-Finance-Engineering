```python
# add an index, converting to appropriate time and  correcting for the missing date.
format = '%Y%m%d%H%M%S%f'
quotes[ 'Time' ] = file_date + quotes[ 'Time' ].astype( str )
times = pd.to_datetime( quotes[ 'Time' ], format = format )
# for now no index, just replace the column
quotes[ 'Time' ] = times
quotes.head()
```

```python
# load trades

trades = pd.read_csv( trade_file, sep = '|',  
                      usecols = [ 'Time', 'Symbol', 'Trade Price', 'Trade Volume', 'Sale Condition' ], 
                                  nrows = 2000 )
format = '%Y%m%d%H%M%S%f'
trades[ 'Time' ] = file_date + trades[ 'Time' ].astype( str )
times = pd.to_datetime( trades[ 'Time' ], format = format )
trades[ 'Time' ] = times
#trades.set_index( times, inplace = True )
#trades = trades.drop( [ 'Time' ], axis=1 )
trades
```

```python
# Enrich trade data with quotes

ticker = 'A'
ticker_quotes = quotes[ quotes[ 'Symbol' ] == ticker ]
ticker_quotes.head(10)
ticker_trades = trades[ trades[ 'Symbol' ] == ticker ]


# merge data
pd.merge_asof( ticker_trades, ticker_quotes, 
               on = 'Time', by = 'Symbol', tolerance = pd.Timedelta( '2000ms' ), allow_exact_matches = False )
```
