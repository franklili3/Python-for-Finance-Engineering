## a.b.	High annualized volatility (> 20%), small/large drift

```python
%matplotlib inline

import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# random walk, matrix
start_price = 100.0

# we want to model some trending to the data
drift1 = -3.0e-6
drift2 = 4.0e-5
```

```python
n_ticks = 390

# This is a simplistic proxy to scale intraday volatility. 
# We'll look at actually measuring it later in the course
volatility = 0.30/np.sqrt(252*390) 

# let's make a series of noise
moves = np.random.normal(0, volatility, n_ticks)

# convert it into a Series
ts = pd.Series(moves)

# throw out the first "noise" value and set our stock price to this.
ts.iloc[0] = start_price

# add a 1.0 to our random walk series
ts = ts.add(1.0)

# add our scalar drift parameter
ts1 = ts.add(drift1)
ts2 = ts.add(drift2)

# calc the series
ts1 = ts1.cumprod()
ts2 = ts2.cumprod()

# a simple plot of the results
# this can be made into a function and optimized if you expect to re-use.

## prepare a 2x1 space on which to plot some data.
figure, plots = plt.subplots(2,1, figsize=(16,9))

## plot addresses are [row,column].  Therefore, plot[1,0] is lower-left.
plots[0].plot(ts1)
plots[0].set_title('Drift: '+ str(drift1))

plots[1].plot(ts2)
plots[1].set_title('Drift: '+ str(drift2))

plt.show()
```

## c.Recreate the basic simulation as in class

```python
# given macd_diff, can we turn that into a series of signals?
macd_diff.head()
```

```python
def calc_signal(x):
    if x > 0:
        return 1 # buy
    elif x < 0:
        return -1 # sell
    else:
        return 0
```

```python
signal = macd_diff.apply(calc_signal)
```

```python
# now let's accumulate some P&L
position = 0
trade_size = 1 # 1 share
pnl = 0
previous_signal = 0
start_price = 0
negative_pnl = 0
positive_pnl = 0
for index, row in stock_data.iterrows():
    current_signal = signal.loc[index]
    if current_signal != previous_signal: # we need to do a trade
        # first check if we have an open position
        if position != 0: 
            # first we need to close the position and calculate P&L
            current_pnl = position * (row['open'] - start_price)
            if current_pnl < 0:
                negative_pnl += 1
            else:
                positive_pnl += 1
            pnl += current_pnl
            msg = "{0} Closing. signal: {1:d} current_price: {2:.2f} start_price: {3:.2f} pnl: {4:.2f}".format(index, current_signal, row['open'], start_price, current_pnl)
            print(msg)
            
        # now open a new position
        position = trade_size * current_signal # buy if signal is positive, sell if signal is negative
        if start_price == 0:
            origin_price,  start_price = row['open']
        else:
            start_price = row['open']
        msg = "{0} Opening. signal: {1:d} current_price: {2:.2f}".format(index, current_signal, row['open'])
        print(msg)    
        previous_signal = current_signal
win_rate = positive_pnl / (positive_pnl + negative_pnl) * 100

print("Final P&L: {0:.2f}".format(pnl))
print('Win rate: {0:.2f}'.format(win_rate))
```

```python
share_minute_logret = np.log(stock_data.close[1:] / stock_data.close[:-1])
strategy_minute_logret = signal.shift(1) * share_minute_logret
strategy_annual_return = (1 + np.exp(strategy_minute_logret.sum())-1) * 252 - 1
```
